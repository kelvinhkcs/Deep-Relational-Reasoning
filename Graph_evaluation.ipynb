````python
import random as ran
import copy
import numpy as np
import jacinle.random as random
import json
import openai
from sklearn.metrics import f1_score
import os
from openai import AzureOpenAI, OpenAI
import graph

n = 40 #adjustable to 10 and 20
Data_prefix = f'Data_{n}/Graph/'
run = 100
file_name = 'Graph'
````

````python
def Graph_GPT4o(name, n, run):
    if name == 'shortest':
        prefix = f'GPT_4o_{n}/Shortest/' 
    elif name = 'connect':
        prefix = f'GPT_4o_{n}/Connectivity/'
    else:
        print('Error in name!')
        return 0
    
    F1 = []
    Data_prefix = f'Data_{n}/Graph/'

    for itr in range(0,run):

        edge_arr = np.load(Data_prefix+file_name+'_edge_graph_'+str(itr)+'.npy')

        relations = graph.generate_relations(n, edge_arr)
        sys_prompt = graph.create_system_prompt()
        
        if name == 'shortest':
            shortest_arr = np.load(Data_prefix+file_name+'_shortest_graph_'+str(itr)+'.npy')
            user_prompt_shortest = graph.create_user_prompt_shortest(n, relations)
        elif name = 'connect'
            connect_arr = np.load(Data_prefix+file_name+'_connect_graph_'+str(itr)+'.npy')
            user_prompt_connect = graph.create_user_prompt_connectivity(n, relations)

        client = graph.create_client_ds('YOUR API KEY',"https://api.deepseek.com")
        llm_response = graph.prompt_llm_ds(client, sys_prompt, eval('user_prompt_'+name))

        try:
            json_output, output_matrix = graph.extract_json_and_numpy_array(llm_response)
        except:
            try:
            response = llm_response.choices[0].message.content
            response = response.replace("```json", "")
            response = response.replace("```", "")
            json_output =  {'content': response}
            json_output = json.dumps(json_output)
            except:
                pass

        try:
            f1 = graph.cal_f1_score(eval(name+'_arr'), output_matrix, weight_method='micro')
            json_output.update({'F1': f1})
            F1.append(f1)
        except:                   
            pass 

        try:
            with open(prefix + file_name + '_' + str(itr) + '_' + name + '.json', 'w', encoding='utf-8') as f:
            json.dump(json_output, f, ensure_ascii=False, indent=4)
        except:                   
            pass
        
````
